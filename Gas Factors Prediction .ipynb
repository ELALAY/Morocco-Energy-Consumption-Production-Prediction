```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Gas Factors.csv', header = 1)
df
```

```python
years = df['Years']
df = df.drop(columns = ['Years'])
df.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
years = np.array(years).reshape(-1, 1)
```

```python
#Splitting Training and Testing Data
X_train = years[:-10]
X_test = years[-10:]
y_train = df[:-10]
y_test = df[-10:]
#X_train
```

**Linear Regression - All Factors**

```python
Linear_model = LinearRegression()
Linear_model.fit(X_train, y_train)
```

```python
predicted_Linear = Linear_model.predict(X_test)
```

```python
predicted_Linear = pd.DataFrame(predicted_Linear)
predicted_Linear
```

**Linear Regression - Predicting by Categories of Factors**

```python
#creating a list of years to predict factors
pred_years = []
for i in range(2020, 2030):
    pred_years.append(i)

pred_years = np.array(pred_years).reshape(-1, 1)
pred_years
```

```python
#Category 1
tr = df['Transportation (TJ)']
ind = df['Industry  (TJ)']
res = df['Residential (TJ)']
com = df['Commercial (TJ)']
cat1 = pd.concat([tr, ind, res, com], axis = 1)
#cat1

cat1_Linear_model = LinearRegression()
cat1_Linear_model.fit(years, cat1)
cat1_predicted_Linear = cat1_Linear_model.predict(X_test)
cat1_predicted_Linear = pd.DataFrame(cat1_predicted_Linear)
cat1_predicted_Linear
```

```python
#Category 2
pop_growth= df['Population growth (%)']
pop_density = df['Population density (%)']
gdp_billions = df['GDP (billion USD)']
cat2 = pd.concat([pop_growth, pop_density, gdp_billions], axis = 1)
#cat2

cat2_Linear_model = LinearRegression()
cat2_Linear_model.fit(years, cat2)
cat2_predicted_Linear = cat2_Linear_model.predict(X_test)
cat2_predicted_Linear = pd.DataFrame(cat2_predicted_Linear)
cat2_predicted_Linear
```

```python
#Category 3
installed_cap_re= df['Installed capacity of renewable energies (MW)']
reserves = df['Reserves (TJ)']
oil_prod_import = df['Imports (TJ)']
cat3 = pd.concat([installed_cap_re, reserves, oil_prod_import], axis = 1)
cat3
cat3_Linear_model = LinearRegression()
cat3_Linear_model.fit(years, cat3)
cat3_predicted_Linear = cat3_Linear_model.predict(X_test)
cat3_predicted_Linear = pd.DataFrame(cat3_predicted_Linear)
cat3_predicted_Linear
```

```python
#Category 4
#Population growth (%).1	GDP (Trillion USD)	Oil prices ($/MW)
pop_growth_world= df['Population growth (%)']
gdp_trillion = df['GDP (Trillion USD)']
oil_prices = df['Natural Gas prices ($/MW)']
cat4 = pd.concat([pop_growth_world, gdp_trillion, oil_prices], axis = 1)
cat4
cat4_Linear_model = LinearRegression()
cat4_Linear_model.fit(years, cat4)
cat4_predicted_Linear = cat4_Linear_model.predict(X_test)
cat4_predicted_Linear = pd.DataFrame(cat4_predicted_Linear)
cat4_predicted_Linear
```

```python
#Category 5
#Energy Security (%)	Energy Equity (%)	Environmental Sustainability (%)
Enery_sec= df['Energy Security (%)']
Energy_equity = df['Energy Equity (%)']
env_sus = df['Environmental Sustainability (%)']
cat5 = pd.concat([Enery_sec, Energy_equity, env_sus], axis = 1)
cat5
cat5_Linear_model = LinearRegression()
cat5_Linear_model.fit(years, cat5)
cat5_predicted_Linear = cat5_Linear_model.predict(X_test)
cat5_predicted_Linear = pd.DataFrame(cat5_predicted_Linear)
cat5_predicted_Linear
```

```python
#Merging the 5 categories findings

facts = pd.concat([cat1_predicted_Linear, cat2_predicted_Linear, cat3_predicted_Linear,
                    cat4_predicted_Linear, cat5_predicted_Linear], axis = 1)
facts.
```
